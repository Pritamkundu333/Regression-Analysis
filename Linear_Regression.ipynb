```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("ml_1.csv")
df
```

```python
plt.xlabel('area(sq ft)')
plt.ylabel('price($)')
plt.scatter(df.area,df.price, color='red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
areas =[1000,1500,2300,3540,4120,4560,5490,3460,4750,2300,9000,8600,7100]
d = pd.DataFrame(areas,columns = ['area'])
d.head()
```

```python
reg.predict(d)
```

```python
plt.xlabel('area',fontsize = 20)
plt.ylabel('price',fontsize = 20)
plt.scatter(df.area,df.price,color ='red',marker ='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

```python
df1 = pd.read_csv("canada_per_capita_income.csv")
df1
```

```python
fun = linear_model.LinearRegression()
fun.fit(df1[['year']],df1.per_capita_income)
```

```python
data = pd.DataFrame([2020,2021,2022,2023,2024],columns = ['year'])
fun.predict(data)
```

```python
plt.xlabel('years',fontsize = 20)
plt.ylabel('per capita income',fontsize = 20)
plt.scatter(df1.year,df1.per_capita_income, marker ='.',color ='red')
plt.plot(df1.year,fun.predict(df1[['year']]),color ='blue')
```
